```python
import matplotlib.pyplot as plt
import numpy as np

def trapezoidal(x, a, b, c, d):
    return np.where(
        x <= a, 0,
        np.where(
            x < b, (x - a) / (b - a),
            np.where(
                x <= c, 1,
                np.where(
                    x < d, (d - x) / (d - c),
                    0
                )
            )
        )
    )

def triangular(x, a, b, c):
    return np.where(
        x <= a, 0,
        np.where(
            x < b, (x - a) / (b - a),
            np.where(
                x < c, (c - x) / (c - b),
                0
            )
        )
    )

def gaussian(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * (sigma ** 2)))
```

# 1. zmienna

$\mathcal{L_1}$ - Rok produkcji  
$H_1(L_1)$ - \{zabytkowy, klasyczny, stary, współczesny, nowoczesny\}  
$\mathcal{X_1}$ - $[1964, 2020] \in \mathbb{N}$  

# Wzory funkcji przynależności dla zmiennej "Rok produkcji"

## Zabytkowy (funkcja trójkątna)
$\mu_{\text{zabytkowy}}(x) = \begin{cases}
\frac{1970-x}{1970-1964} & \text{jeśli } x \in [1964, 1970] \\
0 & \text{w innym wypadku }\\
\end{cases}$

## Klasyczny (funkcja trapezoidalna)
$\mu_{\text{klasyczny}}(x) = \begin{cases}
\frac{x-1965}{1970-1965} & \text{jeśli } x \in [1965, 1970] \\
1 & \text{jeśli } x \in [1970, 1980]\\
\frac{1985-x}{1985-1975} & \text{jeśli } x \in [1980, 1985] \\
0 & \text{w innym wypadku}  \\
\end{cases}$

## Stary (funkcja trapezoidalna)
$\mu_{\text{stary}}(x) = \begin{cases}
\frac{x-1975}{1980-1975} & \text{jeśli } x \in [1975, 1980] \\
1 & \text{jeśli } x \in [1980,2000] \\
\frac{2005-x}{2005-2000} & \text{jeśli } x \in [2000,2005] \\
0 & \text{w innym wypadku } \\
\end{cases}$

## Współczesny (funkcja trapezoidalna)
$\mu_{\text{współczesny}}(x) = \begin{cases}
\frac{x-2000}{2005-2000} & \text{jeśli } x \in [2000, 2005] \\
1 & \text{jeśli } x \in [2005, 2015] \\
\frac{2020-x}{2020-2015} & \text{jeśli } x \in [2015, 2020] \\
0 & \text{w innym wypadku}
\end{cases}$

## Nowoczesny (funkcja trapezoidalna)
$\mu_{\text{nowoczesny}}(x) = \begin{cases}
\frac{x-2015}{2020-2015} & \text{jeśli } x \in [2015, 2020] \\
1 & \text{jeśli } x \in [2020, 2021] \\
\frac{2022-x}{2022-2021} & \text{jeśli } 2021 < x < 2022 \\
0 & \text{w innym wypadku}
\end{cases}$

```python

x = np.linspace(1964, 2020, 1000, endpoint=True)
zabytkowy = triangular(x, 1963, 1964, 1970)
klasyczny = trapezoidal(x, 1964, 1970, 1980, 1985)
stary = trapezoidal(x, 1980, 1985, 2000, 2005)
wspolczesny = trapezoidal(x, 2000, 2005, 2015, 2020)
nowoczesny = trapezoidal(x, 2015, 2020, 2021, 2022)

plt.figure(figsize=(10, 5))
plt.plot(x, zabytkowy, label='Zabytkowy')
plt.plot(x, klasyczny, label='Klasyczny')
plt.plot(x, stary, label='Stary',)
plt.plot(x, wspolczesny, label='Współczesny')
plt.plot(x, nowoczesny, label='Nowoczesny')
# plt.title('Funkcje przynależności dla różnych typów muzyki')
plt.xlabel('Rok')
plt.ylabel('μ(x)')
plt.xlim(1964, 2020)
plt.yticks([0, 0.5, 1])
plt.text(1963, 1.05, 'Zabytkowy', ha='center')
plt.text(1972.5, 1.05, 'Klasyczny', ha='center')
plt.text(1990, 1.05, 'Stary', ha='center')
plt.text(2010, 1.05, 'Współczesny', ha='center')
plt.text(2019.5, 1.05, 'Nowoczesny', ha='center')

# Remove box/frame
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add arrows to axes
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 1964))

# Add arrows to the ends of the axes
ax.plot(2020.1, 0, ">k", ms=8, clip_on=False)
ax.plot(1964, 1.01, "^k", ms=8, clip_on=False)

plt.grid()
plt.tight_layout()
plt.show()
```
