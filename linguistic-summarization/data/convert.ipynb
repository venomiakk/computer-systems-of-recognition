```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df = df[
    [
        'id',
        'Make',
        'Modle',
        'Generation',
        'Year_from',
        'Year_to',
        'Series',
        'Trim',
        'Body_type',
        # 'number_of_seats',
        # 'length_mm',
        # 'width_mm',
        # 'height_mm',
        # 'wheelbase_mm',
        # 'turning_circle_m',
        'full_weight_kg',
        'max_trunk_capacity_l',
        'drive_wheels',
        'transmission',
        'cylinder_layout',
        'number_of_cylinders',
        'engine_type',
        # 'valves_per_cylinder',
        'cylinder_bore_mm',
        'stroke_cycle_mm',
        'capacity_cm3',
        'maximum_torque_n_m',
        'turnover_of_maximum_torque_rpm',
        'engine_hp',
        'engine_hp_rpm',
        'acceleration_0_100_km/h_s',
        'max_speed_km_per_h',
        'number_of_gears',
        'fuel_tank_capacity_l',
        'mixed_fuel_consumption_per_100_km_l',
        'city_fuel_per_100km_l',
        'highway_fuel_per_100km_l'
        # 'emiussion_standard'
        ]]
df
```

```python
df = df.rename(columns={
    'Modle': 'Model',
    'engine_type':'fuel_type',})
df = df.dropna()
df
```

```python
allowed_fuel_types = ['Diesel', 'Gasoline', 'Hybrid', 'Gas']

df = df[df['fuel_type'].isin(allowed_fuel_types)]

df
```

```python
allowed_transmisions = ['Automatic', 'Manual']

df = df[df['transmission'].isin(allowed_transmisions)]

df
```

```python
drive_wheels_mapping = {
    'Rear wheel drive': 'rwd',
    'Front wheel drive': 'fwd',
    'All wheel drive (AWD)': 'awd',
    'Four wheel drive (4WD)': '4wd'
}

df['drive_wheels'] = df['drive_wheels'].replace(drive_wheels_mapping)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None) 

df.head()
```

```python
print(df['Body_type'].unique())
print(df['drive_wheels'].unique())
print(df['transmission'].unique())
print(df['cylinder_layout'].unique())
print(df['number_of_cylinders'].unique())
print(df['fuel_type'].unique())
print(df['number_of_gears'].unique())
```

```python
# Alternatywne podejÅ›cie dla wielu kolumn
numeric_columns = ['full_weight_kg', 'max_trunk_capacity_l', 'capacity_cm3', 'maximum_torque_n_m', 'engine_hp', 'engine_hp_rpm', 'max_speed_km_per_h', 'fuel_tank_capacity_l']

for col in numeric_columns:
    df[col] = (df[col]
              .astype(str)
              .str.replace(',', '.')
              .astype(float)
              .round()
              .astype(int))
```

```python
df = df.astype({
    'id': 'int',
    'Make': 'string',
    'Model': 'string',
    'Generation': 'string',
    'Year_from': 'int',
    'Year_to': 'int',
    'Series': 'string',
    'Trim': 'string',
    'Body_type': 'string',
    'full_weight_kg': 'int',
    'max_trunk_capacity_l': 'int',
    'drive_wheels': 'string',
    'transmission': 'string',
    'cylinder_layout': 'string',
    'number_of_cylinders': 'int',
    'fuel_type': 'string',
    'cylinder_bore_mm': 'float',
    'stroke_cycle_mm': 'float',
    'capacity_cm3': 'int',
    'maximum_torque_n_m': 'int',
    'turnover_of_maximum_torque_rpm': 'int',
    'engine_hp': 'int',
    'engine_hp_rpm': 'int',
    'acceleration_0_100_km/h_s': 'float',
    'max_speed_km_per_h': 'int',
    'number_of_gears': 'int',
    'fuel_tank_capacity_l': 'int',
    'mixed_fuel_consumption_per_100_km_l': 'float',
    'city_fuel_per_100km_l': 'float',
    'highway_fuel_per_100km_l': 'float'
})
```
